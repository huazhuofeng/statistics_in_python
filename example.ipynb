```python
import pandas as pd

data_dir = "/home/data/fhz/project/KY_inulin_summary_stat"
# 读取 Excel 数据
data = pd.read_excel(f"{data_dir}/data/fig2/Figure 2_pathology.xlsx")

# 添加'suhx'列：如果'group'是"CON"，则为0，否则为1
data['suhx'] = data['group'].apply(lambda x: 0 if x == "CON" else 1)

# 添加'inulin'列：如果'group'是"SuHx-IN-Low"或"SuHx-IN-High"，则为1，否则为0
data['inulin'] = data['group'].apply(lambda x: 1 if x in ["SuHx-IN-Low", "SuHx-IN-High"] else 0)
```
